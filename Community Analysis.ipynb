```python
# load pandas
import pandas as pd
import numpy as np
```

```python

# load in the primary care file
data = pd.read_csv("primary_care_access.csv", index_col="PRACTICE_ZIP")

# look at the first ten rows of the data
data.head(10)
```

```python
#Rearrange the data to count the frequency of primary care in each neighborhood.
count = data['PRACTICE_CITY'].value_counts() 
print(count) 
```

```python
##set size of plots
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [15, 7]

##plot the bar graph
count.plot.bar()
```

```python
#create legend code and set pie chart
import matplotlib.patches as mpatches
pie_chart = count.plot.pie()

#legend information
blue_patch = mpatches.Patch(color='blue', label='Pittsburgh')
orange_patch = mpatches.Patch(color='orange', label='Monroeville')
green_patch = mpatches.Patch(color='green', label='Wexford')

#display legend on pie chart
pie_chart.legend(handles=[blue_patch, orange_patch, green_patch])
```
